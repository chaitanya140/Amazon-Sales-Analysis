```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
#Load the Dataset
df = pd.read_csv('amazon.csv')

df.head()
```

```python
df.info()
```

# Data Cleaning

```python
df.isnull().sum()
```

```python
#Remove the Null Values
df.dropna(inplace = True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
# Replace non-numeric values with NaN
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Optionally, fill NaN values with a default value (e.g., 0 or the mean)
df['rating'].fillna(0, inplace=True) 
```

```python
#Convert rating from object to float64.
df['rating'] = df['rating'].astype('float64')

#Convert discounted price from object to int
df['discounted_price'] = df['discounted_price'].replace('[^\d]', '', regex=True).astype(int)

#Convert actual price from object to int
df['actual_price'] = df['actual_price'].replace('[^\d]', '', regex=True).astype(int)
```

```python
df.info()
```

```python
#Check Duplicated Values
df.duplicated().sum()
```

# Data Visualization

```python
df.columns
```

```python
#1. Find out the Distribution of Ratings.
plt.figure(figsize = (9,9))
sns.histplot(x = 'rating' , data = df)
plt.title('Distribution of Ratings')
plt.show()
```

#The histogram shows a strong positive skew in product ratings, with most ratings clustered between 4.0 and 4.5. This indicates high customer satisfaction and a rarity of negative reviews.

```python
#2. Compare top 10 total sales across different product categories.
top_10 = df.groupby('category')['discounted_price'].sum().nlargest(10).reset_index()
plt.figure(figsize=(14,14))
ax = sns.barplot(x='category', y='discounted_price', data=top_10)
ax.bar_label(ax.containers[0], padding=5)
plt.xticks(rotation=90)
plt.title('Top 10 Total Sales For Different Product Categories')
plt.tight_layout()
plt.show()
```

#The "Home Improvement" category, specifically "Electronics & Accessories," exhibits the highest total sales, significantly outperforming other product categories.  Sales drop sharply after the top two categories, indicating a concentration of revenue within a limited number of high-performing product segments.

```python
#3.Find the Top 5 Products with the most customer reviews.
most_customer_reviews = df.groupby('review_content')['product_name'].count().nlargest(5).reset_index()
plt.figure(figsize=(9, 9))
ax = sns.barplot(x='review_content', y='product_name', data=most_customer_reviews)
ax.bar_label(ax.containers[0])
plt.xticks(rotation=90)
plt.title('Top 5 Products With Most Customer Reviews')
plt.show()
```

```python
#4.Find the correlation between product prices and their ratings.
sns.scatterplot(x = 'actual_price' , y = 'rating' , data = df)
plt.title('Correlation Between Product Prices And Their Ratings')
plt.show()
```

```python
#5.Compare the top 5 products across discounted price.
different_shipping_methods = df.groupby('product_name')['discounted_price'].count().nlargest(5).reset_index()
plt.figure(figsize=(8, 8))
plt.pie(different_shipping_methods['discounted_price'], labels=different_shipping_methods['product_name'], autopct='%1.1f%%')
plt.title('Compare Top 5 Products by Discounted Price')
plt.xticks(rotation=90)
plt.show()
```

```python
#6.Which category contributes the most to total revenue, and how has its sales trend evolved over time.
#6.1Use a bar chart to show the Top Category Ratings.
category_rating = df.groupby('product_name')['rating'].count().nlargest(5).reset_index()

plt.figure(figsize=(9, 9))
ax = sns.barplot(x='product_name', y='rating', data=category_rating)
ax.bar_label(ax.containers[0])
plt.xticks(rotation=90)
plt.title('Top Products With Most Category Ratings')
plt.show()
```

```python
#6.2Line plot showing the relationship between sales and discounted percentage.
sales_trend = df.groupby('discounted_price')['discount_percentage'].count().nlargest(10).reset_index()
sns.lineplot(x = 'discounted_price' , y = 'discount_percentage' , data = sales_trend)
plt.title('Relationship Between Sales And Discounted Percentage of the Products.')
plt.show()
```

# Conclusion

#This reveals a strong positive correlation between product ratings and sales, indicating that customer satisfaction significantly drives revenue.  The "Home Improvement" category, particularly "Water Heaters & Accessories," emerges as the top revenue generator, demonstrating substantial market demand and sales volume, though its sales trend over time needs further investigation.  Furthermore, the top 5 products with the most customer reviews highlight popular items with high engagement, while a comparison of discounted prices across these top performers suggests strategic pricing plays a crucial role in their success.
